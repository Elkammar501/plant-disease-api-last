```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Path to the dataset
data_path = r"C:\Users\Admin\Downloads\archive (4)\plantvillage dataset\color"

# Create lists to store image paths and corresponding labels
image_paths = []
labels = []

# Loop through subdirectories (each subdirectory represents a class)
for class_dir in os.listdir(data_path):
    class_path = os.path.join(data_path, class_dir)
    if os.path.isdir(class_path):
        for img in os.listdir(class_path):
            image_paths.append(os.path.join(class_path, img))
            labels.append(class_dir)

# Convert the data into a DataFrame
data = pd.DataFrame({
    'image_path': image_paths,
    'label': labels
})

# Validate image paths and extensions
valid_extensions = ['.jpg', '.jpeg', '.png']
data = data[data['image_path'].str.lower().str.endswith(tuple(valid_extensions))]
data = data[data['image_path'].apply(os.path.exists)]

# Split the data into train, validation, and test sets
train_data, temp_data = train_test_split(data, test_size=0.2, stratify=data['label'], random_state=42)
val_data, test_data = train_test_split(temp_data, test_size=0.5, stratify=temp_data['label'], random_state=42)

# ImageDataGenerator with augmentation (if needed)
train_datagen = ImageDataGenerator(
    rescale=1.0/255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True
)

val_test_datagen = ImageDataGenerator(
    rescale=1.0/255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True
)
# Batch size
batch_size = 256

# Prepare the training data generator
train_generator = train_datagen.flow_from_dataframe(
    dataframe=train_data,
    x_col='image_path',
    y_col='label',
    target_size=(224, 224),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=True
)

# Prepare the validation data generator
val_generator = val_test_datagen.flow_from_dataframe(
    dataframe=val_data,
    x_col='image_path',
    y_col='label',
    target_size=(224, 224),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=True
)

# Prepare the test data generator
test_generator = val_test_datagen.flow_from_dataframe(
    dataframe=test_data,
    x_col='image_path',
    y_col='label',
    target_size=(224, 224),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False
)
```

```python
from tensorflow import keras
from tensorflow.keras import layers
```

```python
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
```

```python
from tensorflow.keras.applications import MobileNetV2

# Specify the path to the pre-trained weights you want to use.
# If you want to use 'imagenet' weights:
weights_path = 'imagenet'

# Or if you have downloaded weights to a specific path:
# weights_path = '/path/to/your/weights.h5'

base_model = MobileNetV2(weights=weights_path, include_top=False, input_shape=(224, 224, 3))

# عرض ملخص النموذج
base_model.summary()
```

```python
for layer in base_model.layers:
    layer.trainable = False
```

```python
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(256, activation='relu')(x)
predictions = Dense(38, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=predictions)
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6)
# Change the filepath to end with .keras
model_checkpoint = ModelCheckpoint('best_resnet_model.keras', save_best_only=True, monitor='val_loss', mode='min')

callbacks = [early_stopping, reduce_lr, model_checkpoint]
```

```python
history = model.fit(
    train_generator,
    epochs=50,
    validation_data=val_generator,
    callbacks=callbacks,
    verbose=1
)
model.save("final_mobilenet_model.h5")
```

```python
predictions = model.predict(test_generator)
```

```python
loss, accuracy = model.evaluate(test_generator)
print(f"Loss: {loss:.4f}")
print(f"Accuracy: {accuracy:.4f}")
```

```python
import numpy as np
from sklearn.metrics import accuracy_score, classification_report

# تحويل التنبؤات إلى تصنيفات
predicted_classes = np.argmax(predictions, axis=1)

# الحصول على التصنيفات الحقيقية
true_classes = test_generator.classes

# حساب دقة النموذج
accuracy = accuracy_score(true_classes, predicted_classes)
print(f"Accuracy: {accuracy:.4f}")

# عرض تقرير التصنيف
print(classification_report(true_classes, predicted_classes))
```
